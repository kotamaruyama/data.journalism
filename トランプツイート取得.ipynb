```python
#ここで取得する設定します。

c = twint.Config()
c.Username = "realDonaldTrump"
c.Store_csv=True
c.Count=True
c.Since='2019-01-01'
c.Until = '2020-01-01'
c.Format="Tweet id:{id}| Date:{date}|Time:{time}|Tweet:{tweet}"
c.Custom['tweet']=['id','date','time','tweet']
c.Output='2019-2020'
twint.run.Search(c)
```
